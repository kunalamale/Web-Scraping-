```python
products=soup.find_all('div',class_="_3pLy-c")
for single in products:
    title=single.find('div',class_= "_4rR01T")
    price=single.find('div',class_="_30jeq3 _1_WHN1")
    print(title.text,"\t",price.text)
```

```python
# create a CSV file.

file=open("products.csv","w",newline='',encoding="utf-8")
wr=csv.writer(file)
wr.writerow(["product name","product price"])

products=soup.find_all('div',class_="_3pLy-c")
for single in products:
    title=single.find('div',class_= "_4rR01T")
    price=single.find('div',class_="_30jeq3 _1_WHN1")
    wr.writerow([title.text,price.text])
    
file.close()
```
